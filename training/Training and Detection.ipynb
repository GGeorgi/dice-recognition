# 8. Load Train Model From Checkpoint

```python
import os
import tensorflow as tf
from object_detection.utils import label_map_util
from object_detection.utils import visualization_utils as viz_utils
from object_detection.builders import model_builder
from object_detection.utils import config_util
```

```python
# Load pipeline config and build a detection model
configs = config_util.get_configs_from_pipeline_file(files['PIPELINE_CONFIG'])
detection_model = model_builder.build(model_config=configs['model'], is_training=False)

# Restore checkpoint
ckpt = tf.compat.v2.train.Checkpoint(model=detection_model)
ckpt.restore(os.path.join(paths['CHECKPOINT_PATH'], 'ckpt-4')).expect_partial()

@tf.function
def detect_fn(image):
    image, shapes = detection_model.preprocess(image)
    prediction_dict = detection_model.predict(image, shapes)
    detections = detection_model.postprocess(prediction_dict, shapes)
    return detections
```

# 9. Detect from an Image

```python
import cv2 
import numpy as np

category_index = label_map_util.create_category_index_from_labelmap(files['LABELMAP'])

colorList = np.random.uniform(low=0, high=255, size=(len(category_index), 3))

def drawBox(img,lineMultiplier=1,fontScale=5):
    image_np = np.array(img)

    input_tensor = tf.convert_to_tensor(np.expand_dims(image_np, 0), dtype=tf.float32)
    detections = detect_fn(input_tensor)

    num_detections = int(detections.pop('num_detections'))
    detections = {key: value[0, :num_detections].numpy()
                  for key, value in detections.items()}
    detections['num_detections'] = num_detections

    # detection_classes should be ints.
    detections['detection_classes'] = detections['detection_classes'].astype(np.int64)

    label_id_offset = 1
    image_np_with_detections = image_np.copy()
    
    bboxs = detections["detection_boxes"]
    classIndexes = detections["detection_classes"].astype(np.int32)
    classScores = detections["detection_scores"]
    
    threshold = 0.5

    imH, imW, imC = image_np_with_detections.shape

    bboxIdx = tf.image.non_max_suppression(
            bboxs,
            classScores,
            max_output_size=50,
            iou_threshold=threshold,
            score_threshold=threshold,
        )

    if len(bboxIdx) != 0:
        for i in bboxIdx:
            bbox = tuple(bboxs[i].tolist())
            classConfidence = round(100 * classScores[i])
            classIndex = classIndexes[i]+1

            classLabelText = category_index[classIndex]['name']
            classColor = colorList[classIndex-1]
            
            displayText = "{}: {}%".format(classLabelText, classConfidence)

            ymin, xmin, ymax, xmax = bbox

            xmin, xmax, ymin, ymax = (
                xmin * imW,
                xmax * imW,
                ymin * imH,
                ymax * imH,
            )
            xmin, xmax, ymin, ymax = int(xmin), int(xmax), int(ymin), int(ymax)

            cv2.rectangle(
                image_np_with_detections, (xmin, ymin), (xmax, ymax), color=classColor, thickness=5*lineMultiplier
            )
            
            cv2.putText(
                    image_np_with_detections,
                    displayText,
                    (xmin, ymin - 20*lineMultiplier),
                    cv2.FONT_HERSHEY_PLAIN,
                    fontScale=fontScale,
                    color=classColor,
                    thickness=5*lineMultiplier,
                )

            lineWidth = min(int((xmax - xmin) * 0.2), int((ymax - ymin) * 0.2))

            cv2.line(
                image_np_with_detections,
                (xmin, ymin),
                (xmin + lineWidth, ymin),
                classColor,
                thickness=14*lineMultiplier,
            )
            cv2.line(
                image_np_with_detections,
                (xmin, ymin),
                (xmin, ymin + lineWidth),
                classColor,
                thickness=14*lineMultiplier,
            )

            cv2.line(
                image_np_with_detections,
                (xmax, ymin),
                (xmax - lineWidth, ymin),
                classColor,
                thickness=14*lineMultiplier,
            )
            cv2.line(
                image_np_with_detections,
                (xmax, ymin),
                (xmax, ymin + lineWidth),
                classColor,
                thickness=14*lineMultiplier,
            )

            cv2.line(
                image_np_with_detections,
                (xmin, ymax),
                (xmin + lineWidth, ymax),
                classColor,
                thickness=14*lineMultiplier,
            )
            cv2.line(
                image_np_with_detections,
                (xmin, ymax),
                (xmin, ymax - lineWidth),
                classColor,
                thickness=14*lineMultiplier,
            )

            cv2.line(
                image_np_with_detections,
                (xmax, ymax),
                (xmax - lineWidth, ymax),
                classColor,
                thickness=14*lineMultiplier,
            )
            cv2.line(
                image_np_with_detections,
                (xmax, ymax),
                (xmax, ymax - lineWidth),
                classColor,
                thickness=14*lineMultiplier,
            )
            
    return image_np_with_detections
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
IMAGE_PATH = os.path.join(paths['IMAGE_PATH'], 'test', '1.jpg')

img = cv2.imread(IMAGE_PATH)
plt.imshow(cv2.cvtColor(drawBox(img,3,10), cv2.COLOR_BGR2RGB))
plt.show()
```

```python
cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while cap.isOpened(): 
    ret, frame = cap.read()
    img=drawBox(frame,1,5)
    cv2.imshow('object detection',  cv2.resize(img, (width, height)))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
```
